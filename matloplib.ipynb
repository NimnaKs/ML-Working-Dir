```python
days = [1,2,3,4,5]

sleeping =[7,8,6,11,7]
eating = [2,3,4,3,2]
working =[7,8,7,2,2]
playing = [8,5,7,8,13]

plt.plot([],[],color='m', label='Sleeping', linewidth=5)
plt.plot([],[],color='c', label='Eating', linewidth=5)
plt.plot([],[],color='r', label='Working', linewidth=5)
plt.plot([],[],color='k', label='Playing', linewidth=5)

plt.stackplot(days, sleeping,eating,working,playing, colors=['m','c','r','k'])

plt.xlabel('x')
plt.ylabel('y')
plt.title('Stack Plot')
plt.legend()
plt.show()
```

### **Linear Regression**

#####  **Predicting House Prices**

##### **Problem Statement :**

###### **You are provided with a dataset containing information about houses, including features like the size of the house (in square feet) and the number of bedrooms. The goal is to predict the price of a house based on these features using linear regression.**

| Size (sqft) | Bedrooms | Price ($) |
|-------------|----------|-----------|
| 2104        | 3        | 399900    |
| 1600        | 3        | 329900    |
| 2400        | 3        | 369000    |
| 1416        | 2        | 232000    |
| 3000        | 4        | 539900    |

##### **Step 01: Load Dataset**

```python
import pandas as pd

data = {
    "Size (sqft)": [2104, 1600, 2400, 1416, 3000],
    "Bedrooms": [3, 3, 3, 2, 4],
    "Price ($)": [399900, 329900, 369000, 232000, 539900]
}

df = pd.DataFrame(data)
print(df)
```

##### **Step 02: Visualize the Data**

```python
import matplotlib.pyplot as plt

# Plotting Size vs Price
plt.scatter(df['Size (sqft)'], df['Price ($)'], color='blue', label='Size')
plt.xlabel('Size (sqft)')
plt.ylabel('Price ($)')
plt.title('Size vs Price')
plt.legend()
plt.show()
```

###### **Small linearlity exists between size and price**

```python
# Plotting Bedrooms vs Price

plt.scatter(df['Bedrooms'], df['Price ($)'], color='red', label='Bedrooms')
plt.xlabel('Bedrooms')
plt.ylabel('Price ($)')
plt.title('Bedrooms vs Price')
plt.legend()
plt.show()
```

###### **Week linearlity exists between bedroom and price**

##### **Step 03: Prepare the dataset**

```python
# Seperate features and target variable

X = df[['Size (sqft)', 'Bedrooms']].values
y = df['Price ($)'].values
print(y)
```

##### **Step 04: Define and Train the Linear Regression Model**

##### **Using scikit-learn's LinearRegression class:**

```python
## pip install scikit-learn
from sklearn.linear_model import LinearRegression
```

```python
# Create the linear regression model
model = LinearRegression()
model.fit(X, y)
```

```python
theta_0 = model.intercept_
theta_1, theta_2 = model.coef_

print(f"Intercept: (theta_0), {theta_0}")
print(f"Coefficient for Size: (theta_1)",{theta_1})
print(f"Coefficient for Bedrooms: (theta_2)",{theta_2})
```

##### **Step 05: Make Predictions**

##### **Using trainned model make predictions**

```python
# Make Predictions

predicted_price = model.predict([[2500, 3]])
print(f"Predicted Price for 2500 sqft and 3 bedrooms: ${predicted_price[0]:.2f}")
```

```python
# Make Predictions

predicted_prices = model.predict(X)
print('Predicted prices',predicted_prices)
```

##### **Step 06: Evaluate the Model**

##### **Compare the predicted prices with actual prices**

```python
# Compare the predicted prices with the actual prices
comparison_df = pd.DataFrame({
    'Actual Price': y, 
    'Predicted Price': predicted_prices,
    'Difference': y - predicted_prices}
)
print(comparison_df)
```
