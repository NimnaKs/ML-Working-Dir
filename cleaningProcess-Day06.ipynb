### **Handling Missing Values**

1. Use Python dropna function
2. Use Python fillna function
3. Use Interpolation method
4. Use Python replace function

```python
%pip install seaborn

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_sales = pd.read_excel("04 Handling Missing Data using Python_Saless.xlsx")
df_sales
```

```python
df_sales.info()
```

##### **dropna function**

```python
df_sales_nan = df_sales.dropna()
df_sales_nan
```

```python
# Drop rows with all missing values
df_sales_nan_all = df_sales.dropna(how="all")
df_sales_nan_all
```

```python
# Drop rows where we have at least 1 valid value
df_sales_nan = df_sales.dropna(thresh=1)
df_sales_nan
```

```python
df_sales_nan = df_sales.dropna(thresh=3)
df_sales_nan
```

##### **fillna**

```python
df_sales_fillna = df_sales.fillna(0)
df_sales_fillna
```

```python
df_sales.fillna({"Day_Temp":df_sales["Day_Temp"].mean(),
                "No_of_Customers":df_sales["No_of_Customers"].median(),
                "Product_Quality":df_sales["Product_Quality"].mode()[0],
                "Sales":df_sales["Sales"].median()})
```

```python
# ffill: propagate the last valid observation forward to the next valid
df_sales_3 = df_sales.ffill()
df_sales_3
```

```python
# bfill: propagate the next valid observation backward to the last valid
df_sales_3 = df_sales.bfill()
df_sales_3
```

```python
# Lets If we want to fill the Nan with the previous or next column value on same row ,  we use axis
df_sales_3 = df_sales.bfill(axis="columns")
df_sales_3
```

```python
df_sales_3 = df_sales.bfill(limit=1)
df_sales_3
```

#### **Interpolation method**

```python
import pandas as pd
import numpy as np

data = {'A': [1, np.nan, np.nan, 4]}
df = pd.DataFrame(data)

# Linear Interpolation
df_interpolated = df.interpolate()
print(df_interpolated)
```

```python
import pandas as pd
import numpy as np

data = {'A': [1, np.nan, np.nan, 4]}
index = pd.to_datetime(['2020-01-01', '2020-01-02', '2020-01-08', '2020-01-12'])
df = pd.DataFrame(data, index=index)

# Linear Interpolation
df_interpolated = df.interpolate(method='time')
print(df_interpolated)
```

```python
data = {'A': [1, np.nan, np.nan, 4]}
df = pd.DataFrame(data)

# pad/forward fill
df_interpolated = df.interpolate(method="pad")
print(df_interpolated)
```

```python
data = {'A': [1, np.nan, np.nan, 4]}
df = pd.DataFrame(data)

# pad/forward fill
df_interpolated = df.interpolate(method="nearest")
print(df_interpolated)
```

```python
data = {'A': [1, np.nan, np.nan, 4]}
df = pd.DataFrame(data, index = [10,20,30,40])

# pad/forward fill
df_interpolated = df.interpolate(method="index")
print(df_interpolated)
```
