```python
%pip install seaborn

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_sales = pd.read_excel("04 Handling Missing Data using Python_Saless.xlsx")
df_sales
```

```python
df_sales.info()
```

##### **dropna function**

```python
df_sales_nan = df_sales.dropna()
df_sales_nan
```

```python
# Drop rows with all missing values
df_sales_nan_all = df_sales.dropna(how="all")
df_sales_nan_all
```

```python
# Drop rows where we have at least 1 valid value
df_sales_nan = df_sales.dropna(thresh=1)
df_sales_nan
```

```python
df_sales_nan = df_sales.dropna(thresh=3)
df_sales_nan
```

##### **fillna**

```python
df_sales_fillna = df_sales.fillna(0)
df_sales_fillna
```
