##### **Step 01: Load Dataset**

```python
import pandas as pd

data = {
    'Dosage' : [10,15,20,25,30,35,40,45,50,55],
    'Blood Pressure' : [120,125,132,140,145,150,158,165,170,178]
}

df = pd.DataFrame(data)
print(df)
```

##### **Step 02: Visualize the Data**

```python
import matplotlib.pyplot as plt

plt.scatter(df['Dosage'], df['Blood Pressure'], color='red')
plt.title('Dosage vs Blood Pressure')
plt.xlabel('Dosage')
plt.ylabel('Blood Pressure')
plt.show('Dosage vs Blood Pressure')
plt.show()
```

###### **Strong Linear Relationship Between Dosage and Blood Pressure.**

###### **So We can use linear regression model**

##### **Step 03: Prepare the dataset**

```python
X = df[['Dosage']].values
y = df['Blood Pressure'].values
```

##### **Step 04: Define and Train the Linear Regression Model**

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python
theta_0 = model.intercept_
theta_1 = model.coef_[0]

print(f'Intercept (theta_0): {theta_0}')
print(f'Coefficient for Dosage (theta_1): {theta_1}')
```

##### **Step 05: Make Predictions**

```python
predicted_blood_pressure = model.predict([[7.5]])
print(f'Predicted Blood Pressure for Dosage 7.5: {predicted_blood_pressure}')
```

```python
predicted_blood_pressure = model.predict(X)
print(f'Predicted Blood Pressure for Dosage : {predicted_blood_pressure}')
```

##### **Step 06: Evaluate the Model**

```python
comparision_df = pd.DataFrame({
    'Actual': y, 
    'Predicted': predicted_blood_pressure,
    'Error': y - predicted_blood_pressure
})

print(comparision_df)
```
